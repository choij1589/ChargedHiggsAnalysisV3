```python
import ROOT
```

```python
ERA = 2017
MEASURE = "muon"
```

```python
f = ROOT.TFile.Open(f"../results/{ERA}/ROOT/{MEASURE}/fakerate_MC.root")
f.ls()
```

```python
QCD = f.Get("fake rate - (QCD_bjet)"); QCD.SetStats(0)
TT = f.Get("fake rate - (TT_bjet)"); TT.SetStats(0)

QCD.GetYaxis().SetRangeUser(10, 50)
TT.GetYaxis().SetRangeUser(10, 50)
```

```python
c = ROOT.TCanvas("c", "c", 800, 600)
c.cd()
QCD.Draw("colz text e")
c.Draw()
```

```python
c.cd()
TT.Draw("colz text e")
c.Draw()
```

```python
ratio = QCD.Clone("ratio")
ratio.Add(TT, -1)
ratio.Divide(QCD)
c.cd()
ratio.Draw("colz text e")
c.Draw()
```
